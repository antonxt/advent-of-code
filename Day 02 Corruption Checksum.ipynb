```python
file = open("day2sheet.txt", "r")
summ = 0
# Part 1: Sum of max - min for each line
ssheet = [[int(num) for num in line.replace('\n', '').split('\t')] for line in file]
print(ssheet)

for nums in ssheet:
    summ += max(nums) - min(nums)
summ
```

```python
# Part 2: Sum of evenly divisible numbers
import itertools
sum2 = 0
for line in ssheet:
    combos = itertools.permutations(line, 2)
    for x,y in combos:
        if x % y == 0:
            print('Line {}: {} / {} = {}'.format(line, x, y, x / y))
            sum2 += int(x / y)
            break

sum2    
```
