```python
def nodes(file_name):
    nodes = {}
    file = open(file_name, 'r')
    for line in file:
        left, right = line.split(' <-> ')
        node = int(left)
        linked = set(int(s) for s in right.split(','))
        nodes[node] = nodes[node] | linked if node in nodes else linked
    return nodes

def linked(nodes, node=0):
    linked_nodes = nodes.pop(node, set())
    for linked_node in linked_nodes:
        linked_nodes = linked_nodes | linked(nodes, linked_node)
    return linked_nodes
```

```python
nodes_dict = nodes('day12sample.txt')
print(nodes_dict)
print(linked(nodes_dict, 0))
print(nodes_dict)
```

```python
# Part 2 - Number of groups
nodes_dict = nodes('day12.txt')
groups = []
while nodes_dict:
    node = next(iter(nodes_dict.keys()))
    groups.append(linked(nodes_dict, node))
print(len(groups), groups)
```
